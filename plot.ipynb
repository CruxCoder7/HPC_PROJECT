```python
import matplotlib.pyplot as plt

def plot_graphs(data, N):
    num_threads = list(data.keys())
    speedup = [data[threads]['speedup'] for threads in num_threads]
    efficiency = [data[threads]['efficiency'] for threads in num_threads]

    plt.figure(figsize=(12, 5))

    # Plot speedup
    plt.subplot(1, 2, 1)
    plt.plot(num_threads, speedup, marker='o')
    plt.title(f'Speedup for N={N}')
    plt.xlabel('Number of Threads')
    plt.ylabel('Speedup')

    # Plot parallel efficiency
    plt.subplot(1, 2, 2)
    plt.plot(num_threads, efficiency, marker='o')
    plt.title(f'Parallel Efficiency for N={N}')
    plt.xlabel('Number of Threads')
    plt.ylabel('Parallel Efficiency (%)')

    plt.tight_layout()
    plt.show()
```
